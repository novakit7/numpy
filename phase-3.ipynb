```python
import numpy as np

marks = np.array([
    [85, 90, 78],
    [72, 88, 91],
    [95, 89, 93],
    [60, 75, 70],
    [88, 92, 85]
])
```

Rows = Students
Columns = Math, Physics, Chemistry

```python
print("shape of the data ", np.shape(marks))

#marks of third student.

print("Marks of third student:", marks[2])
```

```python
print("marks of physics : ", marks[:, 1])
```

```python
# total marks of each student
print("total  marks of all student",  np.sum(marks[:, :]))

print("Total marks of each student:", np.sum(marks, axis=1)) # axis=0 → operate down columns  axis=1 → operate across rows
total_marks_subjetwise = np.sum(marks, axis=0)
print(total_marks_subjetwise)
```

```python
#Find the average marks of each student.
print("average marks of the students", np.average(marks,axis=1))
```

```python
#Find the highest mark in the entire dataset.

print("heighest marks in db", np.max(marks))

print("Highest mark in each subject: ", np.max(marks, axis=0))

print("lowest mark in the entire dataset: ", np.min(marks))
```

```python
#Find the average marks subject-wise.
print("the average marks subject-wise: ", np.mean(marks, axis=0))

# Find students whose Math marks are greater than 80.#
print(marks[marks[:, 0] > 80])
```

```python
#Add 5 grace marks to every Chemistry score.
print(marks)
marks[:, 2] += 5
print("added grace marks  to chemistry: ",marks)
```
