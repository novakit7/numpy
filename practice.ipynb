```python
import numpy as np

marks = np.array([78, 92, 85, 96, 88])
```

```python
lowest_index = np.argmin(marks)

print(lowest_index)

topper_index = np.argmax(marks)

print(topper_index)
```

```python
import numpy as np
marks = np.array([65, 80, 95, 70, 85, 95])

print("Topper index:", np.argmax(marks))
print("Lowest scorer index:", np.argmin(marks))
print("Passed indices:", np.where(marks >= 75)[0])
print("Unique marks:", np.unique(marks))
print("Grace marks:", marks + 5)

extra_marks = np.array([90, 88])
print("Merged:", np.concatenate((marks, extra_marks)))
```
